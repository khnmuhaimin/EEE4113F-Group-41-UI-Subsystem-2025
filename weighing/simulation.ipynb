This simulation models an idealized rod (massless) in continuous contact with the ground, supporting a point mass at its top. The mass can rise or fall, converting potential energy into kinetic and ground reaction forces. This setup enables modeling of energy transfer and transient force responses as the body moves, useful for simulating weight signal dynamics.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.001  # simulation timestep
g = 9.807  # gravitational acceleration

m = 1  # object mass
v0 = 0  # initial velocity of mass
h0 = 0 # initial height of mass

t_stop = 50  # simulation stop time
t = np.arange(0, t_stop, dt)
```

```python
t_jump_stop = 10
F_jump = -np.sin(2*np.pi*(1/(2*t_jump_stop))*t)
F_jump[t > t_jump_stop] = 0
F_obj_on_gnd = F_jump - m * g
plt.plot(t, F_obj_on_gnd)
```
